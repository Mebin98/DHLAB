```python
import numpy as np
import torch
import torchOptics.optics as tt
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')  # Ignore warnings
```

```python
path = '/home/mzmighty17/unknown.npy'
npy_field = np.load(path)
```

```python
timg = torch.tensor(np.load('/home/mzmighty17/unknown.npy')).permute(2,0,1)
# Why permute(2,0,1)?
# Image data at numpy -> generally(height,weight,channel) formation
# At pytorch image data usually be handled as (channel,height,width)
```

```python
tt.show(timg, figsize=(4,4))
```

```python
meta_two = {'wl':(638e-9, 515e-9, 450e-9), 'dx': (2e-6, 2e-6)}
propagate_timg = tt.simulate(tt.Tensor(timg, meta_two),30e-3)
tt.show(propagate_timg.abs(), figsize=(4,4))
```
